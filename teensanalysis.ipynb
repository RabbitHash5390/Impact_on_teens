```python
import pandas as pd
df_habits = pd.read_csv("student_habits_performance.csv") 
df_addiction =pd.read_csv("students social media addiction.csv")
df_suicide = pd.read_csv("social-media-impact-on-suicide-rates.csv") # Change 'your_filename.csv' to your actual file name
```

```python
print(df_habits.head())
print(df_addiction.head())
print(df_suicide.head())

df_addiction.rename(columns={'Student_ID':'student_id','Age': 'age', 'Gender': 'gender','Sleep_Hours_Per_Night': 'sleep_hours',
    'Mental_Health_Score': 'mental_health_score'}, inplace=True)

df_suicide.rename(columns={'sex': 'gender'}, inplace=True)
```

```python
merged_df1_2 = pd.merge(
    df_habits,
    df_addiction, 
    on='student_id',
            suffixes=('_habits', '_addiction'))
```

```python
merged_df1_2['sleep_hours_avg'] = merged_df1_2[['sleep_hours_habits', 'sleep_hours_addiction']].mean(axis=1)
merged_df1_2['mental_health_avg'] = merged_df1_2[['mental_health_rating', 'mental_health_score']].mean(axis=1)
```

```python
print(merged_df1_2.columns)

# Step 1: Create a single 'gender' column from one of the renamed columns
merged_df1_2['gender'] = merged_df1_2['gender_habits'].str.title()

# Step 2: Filter out 'Other' if needed
filtered_df = merged_df1_2[merged_df1_2['gender'] != 'Other']

# Step 3: Standardize casing in df_suicide as well
df_suicide['gender'] = df_suicide['gender'].str.title()

# Step 4: Merge with df_suicide on 'gender'
final_merged = pd.merge(
    filtered_df,
    df_suicide,
    on='gender',
    how='left'
)

print(merged_df1_2.columns.tolist())
```

```python
# STEP 6: Save the final merged DataFrame
final_merged.to_csv("merged_student_socialmedia_data.csv", index=False)
print("Merged dataset saved as 'merged_student_socialmedia_data.csv'")
```

```python
# Rename the duplicated 'sleep_hours' columns to unique names
merged_df1_2.columns = ['sleep_hours_habits' if x=='sleep_hours' and i==0 else
                       'sleep_hours_addiction' if x=='sleep_hours' and i==1 else x
                       for i, x in enumerate(merged_df1_2.columns)]
```

```python
print(merged_df1_2.columns.tolist())
```

```python
# Create a new list of column names
new_cols = []
seen = {}

for col in merged_df1_2.columns:
    if col == 'sleep_hours':
        if col not in seen:
            new_cols.append('sleep_hours_habits')
            seen[col] = 1
        else:
            new_cols.append('sleep_hours_addiction')
    else:
        new_cols.append(col)

merged_df1_2.columns = new_cols
```
